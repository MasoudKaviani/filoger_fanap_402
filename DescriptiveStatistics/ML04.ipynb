```python
import pandas as pd
data = pd.read_csv('flights.csv')
data.head()
```

```python
data['target'] = 0
data
```

```python
data.loc[data['arr_delay'] > 10, 'target'] = 1
```

```python
data
```

```python
data.columns
```

```python
X = data[['month', 'day', 'dep_time', 'sched_dep_time', 'dep_delay',
       'sched_arr_time', 'arr_delay', 'carrier', 'origin', 'dest', 'air_time', 'distance', 'hour', 'minute', 'name']]
y = data['target']
X.head()
```

```python
X = pd.concat([X, pd.get_dummies(X['carrier'])], axis=1)
X.head()
```

```python
X = pd.concat([X, pd.get_dummies(X['origin'])], axis=1)
X.head()
```

```python
X = pd.concat([X, pd.get_dummies(X['dest'])], axis=1)
X.head()
```

```python
X = pd.concat([X, pd.get_dummies(X['name'])], axis=1)
X.head()
```

```python
X = X.drop(['carrier', 'origin', 'dest', 'name'], axis=1)
```

```python
X
```

```python
X.info()
```

```python
X = X.fillna(X.mean())
```

```python
X = (X - X.min()) / (X.max() - X.min())
```

```python
X_train = X.iloc[:250000]
X_test = X.iloc[250000:]
y_train = y.iloc[:250000]
y_test = y.iloc[250000:]
```

```python
X_test.isnull().sum().sum()
```

```python
X_train.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)
```

```python
%%time
y_pred = knn.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, y_pred))
```

```python
metrics.confusion_matrix(y_test, y_pred)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
metrics.classification_report(y_test, y_pred)
```
