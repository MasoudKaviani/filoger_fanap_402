```python
from catboost import CatBoostRegressor
```

```python
model = CatBoostRegressor(iterations=200, learning_rate=0.05, depth=6, eval_metric='MAE', verbose=10, task_type='GPU', save_snapshot=True, snapshot_file='masoud', snapshot_interval=10)
```

```python
model = model.load_model('catboost_reg.h5')
```

```python
model
```

```python
import pandas as pd
data_test = pd.read_csv('test.csv')
data_test.head()
```

```python
data_test.drop(['id'], axis=1, inplace=True)
```

```python
model.predict(data_test)
```

```python
import numpy as np
np.exp(model.predict(data_test))
```
